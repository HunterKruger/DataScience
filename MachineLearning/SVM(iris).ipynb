```python
import numpy as np
import matplotlib.pyplot as plt
from itertools import cycle
from sklearn import svm, datasets
from sklearn.metrics import roc_curve, auc
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import label_binarize
from sklearn.multiclass import OneVsRestClassifier
from scipy import interp

%matplotlib inline
```

```python
iris = datasets.load_iris()
X = iris.data
y = iris.target
```

```python
X
```

```python
y
```

```python
y = label_binarize(y, classes=[0, 1, 2])
y
```

```python
n_classes = y.shape[1]  #count cols of y, which is the num of class
n_classes
```

```python
n_samples, n_features = X.shape
print(n_samples)
print(n_features)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.50,random_state=0)
```

```python
#create model and fit 
random_state = np.random.RandomState(0)
model = OneVsRestClassifier(svm.SVC(kernel='linear', probability=True,random_state=random_state))
clt = model.fit(X_train, y_train)
```

```python
clt.score(X_train, y_train)
```

```python
clt.score(X_test,y_test)
```

```python
y_preds_scores=clt.decision_function(X_test)
y_preds_scores[:5]  # watch classification socres in 5 test samples
```

```python
np.argmax(clt.decision_function(X_test), axis=1)[:5]  # watch classification result in 5 test samples
```

```python
clt.predict_proba(X_test)[:5]   # watch prob of prediction for each class
```

```python
np.argmax(clt.predict_proba(X_test),axis=1)[:5]
```
