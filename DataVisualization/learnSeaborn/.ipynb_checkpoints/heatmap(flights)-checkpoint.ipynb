```python
import pandas as pd 
import numpy as np
import seaborn as sns
from matplotlib import pyplot as plt
%matplotlib inline
#https://seaborn.pydata.org/generated/seaborn.heatmap.html?highlight=heatmap#seaborn.heatmap
```

## random values example

```python
data = np.random.rand(12,15)  #random 2d-array, size(12,15)
sns.heatmap(data) 
```

```python
sns.heatmap(data,cmap = sns.cm.rocket_r,annot=True)  #annot show values #sns.cm.rocket_r inverses colors
```

```python
sns.heatmap(data,cmap = sns.cm.rocket_r,vmax=2,vmin=0)  #set range
```

```python
sns.heatmap(data,cmap = sns.cm.rocket_r,vmax=0.5,vmin=0)  #set range
```

## offical site example

```python
flights = sns.load_dataset("flights")
flights = flights.pivot("month", "year", "passengers") #y-axis, x-axis, bar
sns.heatmap(flights,cmap = sns.cm.rocket_r)
```

```python
sns.heatmap(flights,cmap = sns.cm.rocket_r,annot=True)
```

```python
sns.heatmap(flights,cmap = sns.cm.rocket_r,annot=True,fmt='d') #fmt: format values in Int
```

```python
sns.heatmap(flights,cmap = sns.cm.rocket_r,linewidths=0.9) #linewidths:separate blocks
```

```python
sns.heatmap(flights,cmap = sns.cm.rocket_r,linewidths=0.9, annot=True, fmt='d')
```

```python
sns.heatmap(flights,cmap = 'summer', annot=True, fmt = 'd')
```

```python
#Center the colormap at a specific value:
sns.heatmap(flights, cmap ='coolwarm', annot=True, fmt = 'd', center=flights.loc['June',1954])
```

```python
#Center the colormap at a specific value:
sns.heatmap(flights, cmap ='coolwarm', annot=True, fmt = 'd', center=flights.loc['December',1959])
```

```python
#close bar
sns.heatmap(flights, cmap ='coolwarm', cbar=False)
```
