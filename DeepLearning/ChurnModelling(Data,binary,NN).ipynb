```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Flatten,Dense
```

```python
%matplotlib inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
```

```python
dataset = pd.read_csv('Churn_Modelling.csv')
dataset.head(5)
```

```python
X = dataset.drop(labels=['CustomerId','Surname','RowNumber','Exited'],axis=1)
X.head()
```

```python
y = dataset['Exited']
y.head(5)
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
label1 = LabelEncoder()
X['Geography']=label1.fit_transform(X['Geography'])
X.head(10)
```

```python
label2 = LabelEncoder()
X['Gender']=label1.fit_transform(X['Gender'])
X.head(10)
```

```python
X = pd.get_dummies(X, drop_first=True, columns=['Geography'])
X.head(10)
```

```python
X = pd.get_dummies(X, drop_first=True, columns=['Gender'])
X.head(10)
```

## Feature standardization

```python
from sklearn.preprocessing import StandardScaler
```

```python
X_train, X_test, y_train, y_test = train_test_split(X ,y, test_size=0.2, random_state=0, stratify=y)
```

```python
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.fit_transform(X_test)
```

```python
X_train
```

```python
X_test
```

## Build ANN

```python
model = Sequential()
model.add(Dense(X.shape[1],activation='relu',input_dim=X.shape[1]))  
#1st hidden layer with 11 nodes & input layer with 11 nodes (11 features)
model.add(Dense(128,activation='relu'))  #one hidden layer, 128 units
model.add(Dense(1,activation='sigmoid')) #output layer, binary-classification
```

```python
model.compile(optimizer='adam',loss='binary_crossentropy',metrics=['accuracy'])
```

```python
model.fit(X_train, y_train.to_numpy(), batch_size=10, epochs=10, verbose=1)
```

## Evaluate

```python
y_pred = model.predict_classes(X_test)
```

```python
model.evaluate(X_test,y_test.to_numpy())  # loss: 0.2925  accuracy: 0.8490
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score
```

```python
confusion_matrix(y_test, y_pred)
```

```python
accuracy_score(y_test,y_pred)
```
