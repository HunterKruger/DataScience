```python
import pandas as pd 
dataIris=pd.read_csv('iris.csv')
dataIris.head(5)
```

```python
from sklearn.utils import shuffle
dataIris=shuffle(dataIris)
dataIris.head(5)
```

```python
i=int(len(dataIris)*0.66)      #split data into training set and test set
trainingSet=dataIris.head(i).copy()
testSet=dataIris.tail(len(dataIris)-i).copy()
```

```python
import math
def euclideanDistance(x1,x2,length):
    distance=0
    for i in range(length):
        distance+=(x1[i]-x2[i])**2
    return math.sqrt(distance)
```

```python
import operator 
def getNeighbors(trainingSet,testInstanceArray,k):
    trainingSet["distance"]=""
    trainingSetArray=trainingSet.values
    length=len(testInstanceArray)-1
    for i in range(len(trainingSet)):
        dist=euclideanDistance(testInstanceArray,trainingSetArray[i],length)
        trainingSet.iloc[i,5]=dist
    trainingSet.sort_values(by="distance",ascending=True,inplace=True)
    neighbors=trainingSet.iloc[:k,:].copy()
    return neighbors

#demo
# testInstanceArray=[6.3, 3.3, 6.0, 2.5, 'Virginica']
# neighbors=getNeighbors(trainingSet,testInstanceArray,5)
# neighbors
```

```python
def vote(neighbors):
    classVotes={}
    for i in range(len(neighbors)):
        response=neighbors.iloc[i,-2]
        if response in classVotes:
            classVotes[response]+=1
        else:
            classVotes[response]=1
    sortedVotes=sorted(classVotes.items(),key=operator.itemgetter(1),reverse=True)
    return sortedVotes[0][0]
#demo
# print(vote(neighbors))
```

```python
def getAccuracy(testSet,predictions):
    correct=0
    testSetArray=testSet.values
    for i in range(len(predictions)):
        if(testSetArray[i][-1] == predictions[i]):
            correct+=1
    return (correct/float(len(testSet)))*100
#demo
# predictions=['Versicolor','Versicolor','Versicolor','Versicolor','aaa']
# accuracy=getAccuracy(testSet.iloc[:5,:],predictions)
# print(accuracy)
```

```python
#run!
print("Test set:" + str(len(testSet)))
print("Training set: "+str(len(trainingSet)))
#print(testSet)
testSetArray=testSet.values
#print(testSetArray)
predictions=[]
for i in range(len(testSet)):
    neighbors=getNeighbors(trainingSet,testSetArray[i],3)
    result=vote(neighbors)
    predictions.append(result)
    print(">Result:"+ str(result)+"  >Testset:"+str(testSet.iloc[i,-1]))
accuracy=getAccuracy(testSet,predictions)
print("Accuracy: "+str(accuracy))
```
