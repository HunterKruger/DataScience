```python
import pandas as pd 
from matplotlib import pyplot as plt
```

```python
plt.gcf()  #get current figure
plt.gca()  #get current axis
```

```python
plt.style.use('seaborn')
```

```python
df = pd.read_csv('dev2.csv')
df.columns
```

```python
ages = df['Age']
devSalaries = df['All_Devs']
jsSalaries = df['JavaScript']
pySalaries = df['Python']
```

```python
fig, (ax1,ax2) = plt.subplots(nrows=2,ncols=1)  #init subplots: 2 subplots, share the same X-axis
print(ax1)
print(ax2)
```

## Show 2 subplots in the same window

```python
fig, (ax1,ax2) = plt.subplots(nrows=2,ncols=1,sharex=True)   # sharex=True: share the coordinates' number of 2 x-axis
                                                          
ax1.plot(ages, devSalaries, label='All devs',color='#444444',linestyle='--')

ax2.plot(ages, pySalaries, label='Python')
ax2.plot(ages, jsSalaries, label='Javascript')

ax1.legend()
#ax1.set_xlabel('Ages')
ax1.set_ylabel('Median salary (USD)')
ax1.set_title('Median salary (USD) by age')

ax2.legend()
ax2.set_xlabel('Ages')
ax2.set_ylabel('Median salary (USD)')
#ax2.set_title('Median salary (USD) by age')

plt.tight_layout()
```

## Show 2 subplots in 2 windows

```python
fig1, ax1 = plt.subplots()   
fig2, ax2 = plt.subplots()   

                                                          
ax1.plot(ages, devSalaries, label='All devs',color='#444444',linestyle='--')

ax2.plot(ages, pySalaries, label='Python')
ax2.plot(ages, jsSalaries, label='Javascript')

ax1.legend()
ax1.set_xlabel('Ages')
ax1.set_ylabel('Median salary (USD)')
ax1.set_title('Median salary (USD) by age')

ax2.legend()
ax2.set_xlabel('Ages')
ax2.set_ylabel('Median salary (USD)')
ax2.set_title('Median salary (USD) by age')

plt.tight_layout()

#save as png
fig1.savefig('sub1.png')
fig2.savefig('sub2.png')
```
