```python
import pandas as pd
```

```python
employees = pd.DataFrame([[2, 'chris', 1, 15, 2],
                          [6, 'sandy', 2, 5, 7],
                          [7, 'christine', 1, 5, 4],
                          [12, 'aaron', 3, 3, 3]],
                         columns=['id', 'name', 'title_id', 'years', 'department_id'])
```

```python
employees
```

```python
employees = employees.set_index('id')
```

```python
employees
```

```python
titles = pd.DataFrame([[1, 'President'],
                       [2, 'Lead Engineer'],
                       [3, 'Intern']],
                       columns=['id', 'name'])
```

```python
titles
```

```python
titles = titles.set_index('id')
```

```python
titles
```

```python
employees.join(titles,rsuffix='_title')  # name_title, join by employees.id and titles.id
```

```python
# name_title, join by employees.title_id and titles.id
employees_with_title = employees.join(
    titles, on=['title_id'], rsuffix='_title')
```

```python
employees_with_title
```

```python
departments = pd.DataFrame([[1, 'sales'],
                            [2, 'product'],
                            [3, 'technical support'],
                            [4, 'human resources']],
                            columns=['id', 'name'])
departments
```

```python
departments = departments.set_index('id')
departments
```

```python
employees_with_title.join(departments,on=['department_id'],rsuffix='_dept')
```

```python
employees_with_title.join(departments,on=['department_id'],how='inner',rsuffix='_dept')
# inner join ignores NaN
```

```python
employees_with_title.join(departments,on=['department_id'],how='right',rsuffix='_dept')
# take all rows of DF departments
```
