```python
import pandas as pd
import numpy as np
from sklearn import svm
from matplotlib import pyplot as plt
import seaborn as sns
sns.set(font_scale=1.2)
```

## 2 categories: muffin, cupcake; 2 features: sugar, butter

```python
recipes = pd.read_csv('recipes_muffins_cupcakes.csv')
recipes.head(5)
```

```python
sns.lmplot('Sugar','Butter',data=recipes,hue='Type',palette='Set1',fit_reg=False,scatter_kws={'s':70})
```

```python
#inputs
sugarButter=recipes[['Sugar','Butter']].values
typeLabel=np.where(recipes['Type']=='Muffin',0,1)
typeLabel
```

```python
#fit model with C=2^-5   # Low C: soft margin, prioritize simplicity
model=svm.SVC(kernel='linear', C=2**-5)
model.fit(sugarButter,typeLabel)
```

```python
#visualize results

#get hyperplane
w = model.coef_[0]
a = -w[0]/w[1]  #slope
xx = np.linspace(5,30)
yy = a*xx-(model.intercept_[0])/w[1]  # y = slope X - intercept

#get parallels
b = model.support_vectors_[1]
yyDown = a*xx+(b[1]-a*b[0])
b = model.support_vectors_[-1]
yyUp = a*xx+(b[1]-a*b[0])
```

```python
#hyperplane
sns.lmplot('Sugar','Butter',data=recipes,hue='Type',palette='Set1',fit_reg=False,scatter_kws={'s':70})
plt.plot(xx,yy,linewidth=2,color='black')
# margins and support vectors
sns.lmplot('Sugar','Butter',data=recipes,hue='Type',palette='Set1',fit_reg=False,scatter_kws={'s':70})
plt.plot(xx,yy,linewidth=2,color='black')
plt.plot(xx,yyDown,'k--')
plt.plot(xx,yyUp,'k--')
plt.scatter(model.support_vectors_[:,0], model.support_vectors_[:,1],s=80,facecolors='none')
```

```python
#fit model with C=2^7  # High C, priorize making few mistakes
model=svm.SVC(kernel='linear', C=2**7)
model.fit(sugarButter,typeLabel)
```

```python
#visualize results

#get hyperplane
w = model.coef_[0]
a = -w[0]/w[1]  #slope
xx = np.linspace(5,30)
yy = a*xx-(model.intercept_[0])/w[1]  # y = slope X - intercept

#get parallels
b = model.support_vectors_[0]
yyDown = a*xx+(b[1]-a*b[0])
b = model.support_vectors_[-1]
yyUp = a*xx+(b[1]-a*b[0])
```
