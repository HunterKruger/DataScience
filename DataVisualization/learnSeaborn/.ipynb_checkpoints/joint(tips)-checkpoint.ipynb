```python
import pandas as pd 
import numpy as np
import seaborn as sns
from matplotlib import pyplot as plt
%matplotlib inline
#https://seaborn.pydata.org/generated/seaborn.jointplot.html?highlight=joint#seaborn.jointplot
```

```python
tips=sns.load_dataset('tips')
tips.head()
```

```python
iris=sns.load_dataset('iris')
iris.head()
```

```python
sns.jointplot(x='total_bill',y='tip',data=tips)
```

```python
sns.jointplot(x='sepal_length',y='sepal_width',data=iris)
```

```python
sns.jointplot(x='sepal_length',y='sepal_width',data=iris)
```

```python
sns.jointplot(x='total_bill',y='tip',data=tips,kind='reg')
```

```python
sns.jointplot(x='sepal_length',y='sepal_width',data=iris,kind='hex')
```

```python
sns.jointplot(x='sepal_length',y='sepal_width',data=iris,kind='kde',color='red')
```

```python
from scipy.stats import spearmanr
```

```python
sns.jointplot(x='total_bill',y='tip',data=tips,stat_func=spearmanr)
```

```python
sns.jointplot(x='size',y='total_bill',data=tips,stat_func=spearmanr, color='red',ratio=4, height=6)
```

```python
sns.jointplot(x='size',y='total_bill',data=tips,stat_func=spearmanr, color='red',ratio=7, height=10)
```
