```python
import pandas as pd 
import numpy as np
import seaborn as sns
from matplotlib import pyplot as plt
%matplotlib inline
#https://seaborn.pydata.org/generated/seaborn.pairplot.html?highlight=pair%20plot#seaborn.pairplot
```

```python
iris = sns.load_dataset('iris')
iris.head()
```

```python
sns.pairplot(iris)
```

```python
tips = sns.load_dataset('tips')
```

```python
sns.pairplot(tips)
```

```python
sns.pairplot(iris,hue='species',palette='husl', markers=['o','D','s'])
```

```python
sns.pairplot(iris, vars=['sepal_length','sepal_width'],height=4)
```

```python
sns.pairplot(iris, x_vars=['sepal_length','sepal_width'],y_vars=['petal_length','petal_width'],
             hue='species',palette='husl')
```

```python
sns.pairplot(iris,diag_kind='kde', hue='species',palette='husl') #Use kernel density estimates for univariate plots:
```

```python
sns.pairplot(iris, diag_kind="kde", markers="+",
                  plot_kws=dict(s=50, edgecolor="b", linewidth=1), #plot
                  diag_kws=dict(shade=True))  #
```
