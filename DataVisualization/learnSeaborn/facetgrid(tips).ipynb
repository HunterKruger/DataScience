```python
import pandas as pd 
import numpy as np
import seaborn as sns
from matplotlib import pyplot as plt
%matplotlib inline
#https://seaborn.pydata.org/generated/seaborn.FacetGrid.html?highlight=facetgrid
```

```python
tips = sns.load_dataset('tips')
tips.head()
```

```python
x = sns.FacetGrid(tips, row='smoker', col='time')
x = x.map(plt.hist,'total_bill', color ='green', edgecolor="w")
```

```python
x = sns.FacetGrid(tips, row='smoker', col='time')
x = x.map(plt.scatter,'total_bill', 'tip', color ='red', edgecolor="w")
```

```python
x = sns.FacetGrid(tips, hue='smoker', col='time',palette='coolwarm')
x = x.map(plt.scatter,'total_bill', 'tip', edgecolor="w").add_legend()
```

```python
x = sns.FacetGrid(tips, row='smoker', col='time')
x = x.map(sns.regplot,'total_bill', 'tip', color ='blue')
```

```python
x = sns.FacetGrid(tips, col='day')
x = x.map(sns.boxplot, 'total_bill', 'time')
```

```python
x = sns.FacetGrid(tips, col='day',height=4,aspect=.4,col_order=['Sun','Sat','Fri','Thur']) #aspect:长宽比
x = x.map(sns.boxplot,'time', 'total_bill', color='red' )
```
