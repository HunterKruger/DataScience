```python
import pandas as pd 
import numpy as np
import seaborn as sns
from matplotlib import pyplot as plt
from numpy import median
#https://seaborn.pydata.org/generated/seaborn.boxplot.html?highlight=box#seaborn.boxplot
%matplotlib inline
```

```python
sns.set(style="whitegrid")
tips = sns.load_dataset("tips")
```

```python
sns.boxplot(x=tips["total_bill"])
```

```python
sns.boxplot(x="day", y="total_bill", data=tips)
```

```python
sns.boxplot(x="day", y="total_bill", hue="smoker",
            data=tips, palette="Set3")
```

```python
ax = sns.boxplot(x="day", y="total_bill", hue="time",
               data=tips, linewidth=2.5)
```

```python
#Control box order by passing an explicit order:
sns.boxplot(x="time", y="tip", data=tips, order=["Dinner", "Lunch"])
```

```python
iris = sns.load_dataset("iris")
sns.boxplot(data=iris, orient="v", palette="Set2")   #orient="h" is horizonal, vertical by default
```

```python
#Use hue without changing box position or width:
tips["weekend"] = tips["day"].isin(["Sat", "Sun"])
sns.boxplot(x="day", y="total_bill", hue="weekend",
                 data=tips, dodge=False)
```

```python
#Use swarmplot() to show the datapoints on top of the boxes:
sns.boxplot(x="day", y="total_bill", data=tips)
sns.swarmplot(x="day", y="total_bill", data=tips, color=".25")
```
