```python
from matplotlib import pyplot as plt
import pandas as pd 
%matplotlib inline    
plt.style.use('seaborn-poster')
```

```python
df = pd.read_csv('dev2.csv')
```

```python
df.head(5)
```

```python
ages = df['Age']
allDev = df['All_Devs']
pyDev = df['Python']
jsDev = df['JavaScript']
```

## pyDev vs allDev

```python
plt.plot(ages, allDev, color='#444444', linestyle='--', label = 'All devs')
plt.plot(ages, pyDev, label = 'pyDev')
plt.legend()
plt.xlabel('Ages')
plt.ylabel('Median salary (USD)')
plt.title('Median salary (USD) by age')
plt.tight_layout()
plt.fill_between(ages,pyDev,allDev,alpha=0.25)  #pyDev vs allDev; alpha is transparency
```

## pyDev vs overallMedian

```python
plt.plot(ages, pyDev, label = 'pyDev')
plt.xlabel('Ages')
plt.ylabel('Median salary (USD)')
plt.title('Median salary (USD) by age')
plt.tight_layout()
overallMedian=57827
plt.fill_between(ages,pyDev,overallMedian,where=(pyDev>overallMedian),interpolate=True, alpha=0.25,label='Above avg')  
#pyDev vs overallMedian; alpha is transparency
plt.fill_between(ages,pyDev,overallMedian,where=(pyDev<=overallMedian),interpolate=True, alpha=0.25, color='red',label='Below avg')
plt.legend()
```

## pyDev vs allDev

```python
plt.plot(ages, allDev, color='#444444', linestyle='--', label = 'All devs')
plt.plot(ages, pyDev, label = 'pyDev')
plt.xlabel('Ages')
plt.ylabel('Median salary (USD)')
plt.title('Median salary (USD) by age')
plt.tight_layout()
plt.fill_between(ages,pyDev,allDev,where=(pyDev>allDev),interpolate=True, alpha=0.25)  
#pyDev vs allDev; alpha is transparency
plt.fill_between(ages,pyDev,overallMedian,where=(pyDev<=allDev),interpolate=True, alpha=0.25, color='red')
plt.legend()
```
