```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def f(x):
    return x**2 + 10 * np.sin(x)
```

```python
x= np.linspace(-10, 10, 51)
x
```

```python
y = f(x)
y
```

```python
fig = plt.figure(figsize=(5,4), dpi=80)
plt.plot(x, y)
plt.grid('on')
plt.show()
```

```python
import scipy.optimize as opt
```

```python
## find min of f(x) by BFGS metthod
# test1

[x_opt, f_opt, g_opt, B_opt, func_calls, grad_calls, warn_flag]=\
    opt.fmin_bfgs(f, # obj func
                  x0 = 0, # initial guess
                  maxiter= 2000,
                  full_output=True
                 )
```

```python
print(x_opt)
print(f_opt)
```

```python
fig = plt.figure(figsize=(5,4), dpi=80)
plt.plot(x, y)
plt.plot([0,x_opt],[f(0),f_opt],'o')
plt.grid('on')
plt.show()
```

```python
## find min of f(x) by BFGS metthod
# test1

[x_opt, f_opt, g_opt, B_opt, func_calls, grad_calls, warn_flag]=\
    opt.fmin_bfgs(f, # obj func
                  x0 = 6, # initial guess
                  maxiter= 2000,
                  full_output=True
                 )

fig = plt.figure(figsize=(5,4), dpi=80)
plt.plot(x, y)
plt.plot([6,x_opt],[f(6),f_opt],'o')
plt.grid('on')
plt.show()
```

```python
## search the global minimum using basinhopping

res = opt.basinhopping(f,
                       x0 = 6,
                       T = 1,  # the temperature parameter
                       stepsize=2) # initial step size
res
```

```python
fig = plt.figure(figsize=(5,4), dpi=80)
plt.plot(x, y)
plt.plot([6,res.x],[f(6),res.fun],'o')
plt.grid('on')
plt.show()
```
