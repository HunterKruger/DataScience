## Method 2

```python
df = df.select_dtypes(include=['int','float'])
headers = list(df.columns.values)
fields=[]
for field in headers:
    fields.append({
        'name':field,
        'mean':df[field].mean(),
        'var':df[field].var(),
        'std':df[field].std()
    })
for field in fields:
    print(field)
```

```python
df2 = pd.DataFrame(fields)
df2
```
