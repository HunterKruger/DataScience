```python
from matplotlib import pyplot as plt
%matplotlib inline    
plt.style.use('fivethirtyeight')
```

## Simple example

```python
slices=[120,80,30,20]
labels=['A','B','C','D']
colors=['blue','red','yellow','green']
plt.pie(slices,labels=labels,wedgeprops={'edgecolor':'black'})  #use wedgeprops to add edge
plt.title('Pie chart example')
plt.tight_layout()
```

## Popular Language 

```python
labels2=['JavaScript', 'HTML/CSS', 'SQL', 'Python', 'Java']
slices2=[59219,55466,47544,36443,35917]
explode=[0,0,0,0.2,0]  # 0.2 = 20% of radius
```

```python
plt.pie(slices2, labels=labels2, wedgeprops={'edgecolor':'black'},  #use explode to control position
        explode=explode,shadow=True,startangle=90, autopct='%1.1f%%')  #use autopct to show percentage
plt.title('Pie chart example')
plt.tight_layout()
```
