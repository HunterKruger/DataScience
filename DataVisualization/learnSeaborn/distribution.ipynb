```python
import pandas as pd 
import numpy as np
import seaborn as sns
from matplotlib import pyplot as plt
from numpy import median
#https://seaborn.pydata.org/generated/seaborn.distplot.html?highlight=dist#seaborn.distplot
%matplotlib inline
```

```python
num = np.random.randn(150)
#create 150 random values in normal distribution
```

```python
sns.distplot(num,color='red')
```

```python
#add label
label_dist=pd.Series(num, name="var x")
sns.distplot(label_dist) 
```

```python
sns.distplot(label_dist,vertical=True) 
```

```python
sns.distplot(label_dist,hist=False)   #do not show histogram
```

```python
sns.distplot(label_dist,hist=False,rug=True)   #do not show histogram, show rug
```

```python
sns.distplot(label_dist, rug=True, rug_kws={"color": "g"},  #rug
                   kde_kws={"color": "k", "lw": 3, "label": "KDE"},  #straight line fit
                   hist_kws={"histtype": "step", "linewidth": 3,  #histogram
                             "alpha": 1, "color": "g"})
```
