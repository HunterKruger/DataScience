```python
import pandas as pd 
import numpy as np
import seaborn as sns
from matplotlib import pyplot as plt
%matplotlib inline
#https://seaborn.pydata.org/generated/seaborn.kdeplot.html?highlight=kde#seaborn.kdeplot
```

```python
mean = [0,0]
covariance = [[0.2,0],[0,3]]
```

```python
xAxis,yAxis=np.random.multivariate_normal(mean,covariance,size=40).T
```

```python
sns.kdeplot(xAxis,color='green')
```

```python
sns.kdeplot(yAxis,shade=True)
```

```python
sns.kdeplot(xAxis,yAxis,shade=True)
```

```python
sns.kdeplot(xAxis,yAxis,cmap='RdBu')
```

```python
sns.kdeplot(xAxis,yAxis,cmap='coolwarm',n_levels=18)
```

```python
sns.kdeplot(yAxis,shade=True,bw=0.1,vertical=True) #bandwidth
```

```python
iris = sns.load_dataset('iris')
iris
```

```python
setosa = iris.loc[iris.species=='setosa']
virginica = iris.loc[iris.species=='virginica']
```

```python
sns.kdeplot(setosa.petal_length,setosa.petal_width,cmap='coolwarm')
sns.kdeplot(virginica.petal_length,virginica.petal_width,cmap='RdBu',n_levels=14)
```
