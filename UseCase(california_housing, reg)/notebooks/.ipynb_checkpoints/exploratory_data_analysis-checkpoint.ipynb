```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
train = pd.read_csv('../data/raw_data/train.csv')
test = pd.read_csv('../data/raw_data/test.csv')
```

```python
train.shape,test.shape  # no y label in test set, only analyzing training data
```

```python
data = pd.concat([train.drop('SalePrice',axis=1),test],axis=0)
data.shape
# train + test, no y label
```

```python
def data_overview(df):
    print("Rows :  ", df.shape[0])
    print("Columns:  ", df.shape[1])
    print()
    print("Feature types:  ")
    print(df.dtypes)
    print()
    print('Missing values : ')
    print(df.isnull().sum())
    print()
    print('Unique values:')
    print(df.nunique())


def feature_overview(feature):
    print("Type: "+str(train[feature].dtypes))
    print("-------------------------------------------")
    print("Missing values in train: "+str(train[feature].isnull().sum()))
    print("Missing rate in train: " +
          str(train[feature].isnull().sum()/train.shape[0]))
    print("-------------------------------------------")
    print("Unique values in train: "+str(train[feature].nunique()))
    print("Unique values in test: "+str(test[feature].nunique()))
    print("Unique values in all: "+str(data[feature].nunique()))
    if train[feature].dtypes == 'object':
        train[feature].value_counts(dropna=False)


data_overview(data)
```
