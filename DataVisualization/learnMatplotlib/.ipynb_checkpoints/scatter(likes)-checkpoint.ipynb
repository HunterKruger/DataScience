```python
from matplotlib import pyplot as plt
import pandas as pd 
%matplotlib inline  
plt.style.use('seaborn')
```

```python
x=[5,7,8,5,6,7,9,2,3,4,4,4,2,6,3,6,8,6,4,1]
y=[7,4,3,9,1,3,2,5,2,4,8,7,1,6,4,9,7,7,5,1]
```

## Simple example 1

```python
plt.scatter(x,y,s=100,c='green',marker='X',edgecolors='black',linewidths=1, alpha=0.3)  
# s is size of dots  # alpha is transparency
```

## Simple example 2

```python
colors=[7,5,9,7,5,7,2,5,3,7,1,2,8,1,9,2,5,6,7,5]
sizes=[209,486,381,255,191,315,185,228,174,538,239,394,399,153,1273,293,436,501,397,539]
plt.scatter(x,y,s=sizes,c=colors,cmap='Greens', edgecolors='black',linewidths=1)  
# s is size of dots  # alpha is transparency
cbar = plt.colorbar() # set a color bar
cbar.set_label('Satisfaction')
```

## Use pandas

```python
data = pd.read_csv('likes.csv')
data.columns
```

```python
viewCount = data['view_count']
likes = data['likes']
ratio = data['ratio']
```

```python
# ex1
plt.scatter(viewCount,likes,edgecolors='black',linewidths=1)
plt.title('Trending Youtude videos')
plt.xlabel('View count')
plt.ylabel('Total likes')
plt.tight_layout()
```

```python
# ex2
plt.scatter(viewCount,likes,edgecolors='black',linewidths=1,alpha=0.75)
plt.xscale('log')
plt.yscale('log')
plt.title('Trending Youtude videos')
plt.xlabel('View count')
plt.ylabel('Total likes')
plt.tight_layout()
```

```python
# ex3
plt.scatter(viewCount,likes,c=ratio,cmap='summer',edgecolors='black',linewidths=1,alpha=0.75)
plt.xscale('log')
plt.yscale('log')
cbar2=plt.colorbar()
cbar2.set_label('Like/Dislike Ratio')
plt.title('Trending Youtude videos')
plt.xlabel('View count')
plt.ylabel('Total likes')
plt.tight_layout()
```
