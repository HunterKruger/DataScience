```python
import pandas as pd 
import numpy as np
import seaborn as sns
from matplotlib import pyplot as plt
from numpy import median
#https://seaborn.pydata.org/generated/seaborn.stripplot.html?highlight=strip#seaborn.stripplot
%matplotlib inline
```

```python
sns.set(style="whitegrid")
tips = sns.load_dataset("tips")
```

```python
sns.stripplot(x=tips["total_bill"])
```

```python
#Group the strips by a categorical variable:
sns.stripplot(x="day", y="total_bill", data=tips)
```

```python
#use jitter to shake the dots
sns.stripplot(x="day", y="total_bill", data=tips, jitter=True)
```

```python
#Use a smaller amount of jitter:
sns.stripplot(x="day", y="total_bill", data=tips, jitter=0.05)
```

```python
#Draw horizontal strips:
sns.stripplot(x="total_bill", y="day", data=tips,jitter=True)
```

```python
#Draw outlines around the points:
sns.stripplot(x="total_bill", y="day", data=tips,
                    jitter=True, linewidth=1)
```

```python
#Nest the strips within a second categorical variable:
sns.stripplot(x="sex", y="total_bill", hue="day",
                    data=tips, jitter=True)
```

```python
#Draw each level of the hue variable at different locations on the major categorical axis: use dodge
sns.stripplot(x="day", y="total_bill", hue="smoker",
                    data=tips, jitter=True,
                    palette="Set2", dodge=True)
```

```python
#Draw strips with large points and different aesthetics:
sns.stripplot("day", "total_bill", "smoker", data=tips,
                    palette="Set2", size=20, marker="D",
                    edgecolor="gray", alpha=.25)
```

```python
#Use catplot() to combine a stripplot() and a FacetGrid. 
#This allows grouping within additional categorical variables. 
#Using catplot() is safer than using FacetGrid directly, 
#as it ensures synchronization of variable order across facets:

#FacetGrid is multiplot

sns.catplot(x="sex", y="total_bill",
                 hue="smoker", col="time",
                 data=tips, kind="strip",
                 jitter=True,
                 height=4, aspect=.7);
```
