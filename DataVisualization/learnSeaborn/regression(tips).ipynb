```python
import pandas as pd 
import numpy as np
import seaborn as sns
from matplotlib import pyplot as plt
%matplotlib inline
#https://seaborn.pydata.org/generated/seaborn.regplot.html?highlight=regplot#seaborn.regplot
```

```python
tip = sns.load_dataset('tips')
tip.head()
```

```python
sns.regplot(x='total_bill',y='tip',data=tip,color='green')
```

```python
sns.regplot(x='total_bill',y='tip',data=tip,
            scatter_kws={'marker':'D','color':'blue'},line_kws={'color':'red','linewidth':3})
```

```python
sns.regplot(x='size',y='total_bill',data=tip,x_jitter=0.1)  #x_jitter: shake x
```

```python
sns.regplot(x='size',y='total_bill',data=tip,x_estimator=np.mean)  
```

```python
mean=[2,5]
covariance=[[1.1, 0.4],[2.2 ,3]]
xAxis,yAxis=np.random.multivariate_normal(mean,covariance,100).T
```

```python
sns.regplot(x=xAxis,y=yAxis)
```

```python
seriesX=pd.Series(xAxis, name="var x")
seriesY=pd.Series(yAxis, name="var y")
sns.regplot(x=seriesX,y=seriesY,marker='D',color='grey')
```
