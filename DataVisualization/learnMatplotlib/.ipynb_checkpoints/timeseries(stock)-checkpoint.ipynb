```python
import pandas as pd 
from datetime import datetime, timedelta
from matplotlib import pyplot as plt
from matplotlib import dates as mpl_dates
plt.style.use('seaborn')
%matplotlib inline  
```

## Simple example 1

```python
dates = [datetime(2019,5,24),datetime(2019,5,25),datetime(2019,5,26),
         datetime(2019,5,27),datetime(2019,5,28),datetime(2019,5,29),datetime(2019,5,30)]
y=[0,1,3,4,6,5,7,]
```

```python
plt.plot_date(dates,y)
```

```python
plt.plot_date(dates,y,linestyle='solid')
```

```python
plt.plot_date(dates,y,linestyle='solid')
plt.gcf().autofmt_xdate()
```

```python
plt.plot_date(dates,y,linestyle='solid')
plt.gcf().autofmt_xdate()
dateFormat=mpl_dates.DateFormatter('%b,%d %Y')
plt.gca().xaxis.set_major_formatter(dateFormat)
```

## Use pandas

```python
data = pd.read_csv('stock.csv')
data.columns
```

```python
#ex1
priceDate = data['Date']
priceClose = data['Close']
plt.plot_date(priceDate,priceClose,linestyle='solid')
plt.gcf().autofmt_xdate()
plt.title('Bitcoin prices')
plt.xlabel('Date')
plt.ylabel('Closing prices')
plt.tight_layout()
```

```python
#ex2

#handle the 'Date' column in dataframe
data['Date']=pd.to_datetime(data['Date'])  # str -> date
data.sort_values('Date',inplace=True)

#then read columns
priceDate = data['Date']
priceClose = data['Close']

#plot
plt.plot_date(priceDate,priceClose,linestyle='solid')

#format date
plt.gcf().autofmt_xdate()
dateFormat=mpl_dates.DateFormatter('%b,%d %Y')
plt.gca().xaxis.set_major_formatter(dateFormat)

plt.title('Bitcoin prices')
plt.xlabel('Date')
plt.ylabel('Closing prices')
plt.tight_layout()
```
