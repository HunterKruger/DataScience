```python
import pandas as pd 
import numpy as np
import seaborn as sns
from matplotlib import pyplot as plt
from numpy import median
%matplotlib inline
sns.set(color_codes=True)
#https://seaborn.pydata.org/generated/seaborn.clustermap.html?highlight=cluster#seaborn.clustermap
```

## examples from official site

```python
iris = sns.load_dataset("iris")
species = iris.pop("species")  #need to drop the labels 
g = sns.clustermap(iris)
```

```python
#Use a different clustering method:
g = sns.clustermap(iris, method="single")
```

```python
#Use a different colormap and ignore outliers in colormap limits:
g = sns.clustermap(iris, cmap="mako", robust=True)
```

```python
#Change the size of the figure:
g = sns.clustermap(iris, figsize=(6, 7))
```

```python
#Standardize the data within the columns:
g = sns.clustermap(iris, standard_scale=1)
```

```python
#Normalize the data within the rows:
g = sns.clustermap(iris, z_score=0)
```

## other examples

```python
flights = sns.load_dataset('flights')
flights.head()
```

```python
cmap = sns.cm.rocket_r
flights=flights.pivot('month','year','passengers')
```

```python
sns.clustermap(flights,cmap=cmap) #cmap inverses colors
```

```python
sns.clustermap(flights,col_cluster=False,cmap=cmap) #only cluster rows
```

```python
sns.clustermap(flights,row_cluster=False,cmap=cmap) #only cluster cols
```

```python
sns.clustermap(flights,cmap='coolwarm',   #change theme
               linewidths=1,    #separate blocks
               figsize=(8,8))  #adjust size
```

```python
sns.clustermap(flights,cmap='coolwarm',   
               linewidths=1,  
               figsize=(8,8),
              standard_scale=1)    #standarize data across col:1, row:0
```

```python
sns.clustermap(flights,cmap='coolwarm',   
               linewidths=1,  
               figsize=(8,8),
              standard_scale=0)    #standarize data across col:1, row:0
```

```python
sns.clustermap(flights,cmap='coolwarm',   
               linewidths=1,  
               figsize=(8,8),
              z_score=1)    #normalize data by col:1, row :0
```

```python
sns.clustermap(flights,cmap='coolwarm',   
               linewidths=1,  
               figsize=(8,8),
              z_score=0)    #normalize data by col:1, row :0
```
